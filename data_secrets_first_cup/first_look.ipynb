```python
import pandas as pd
```

```python
df_train = pd.read_csv('data/train_target.csv')
df_train
```

```python
df_orders = pd.read_csv('data/orders.csv')
df_orders
```

```python
print('df_train ClientUUId & df_orders ClientUUId\t', len(set(df_train['ClientUUId']) & set(df_orders['ClientUUId'])))
print('df_train ClientUUId\t\t\t\t', len(set(df_train['ClientUUId'])))
print('df_orders ClientUUId\t\t\t\t', len(set(df_orders['ClientUUId'])))
```

```python
df_merge = pd.merge(df_orders, df_train, how='outer', on='ClientUUId')
print(df_merge[['apply_promo_x', 'apply_promo_y']].isna().mean())
df_merge
```
